```python
# import libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# loading my data into a pandas dataframe
df = pd.read_excel('/content/drive/MyDrive/dataset/clean call dataset.xlsx')
df.head(10)
```

```python
df.info()
```

```python
# line chart
avg_speed_over_time = df.groupby("Date")["Speed of answer in seconds"].mean()
plt.plot(avg_speed_over_time.index, avg_speed_over_time.values)
plt.xlabel("Date")
plt.ylabel("Average Speed of Answer")
plt.title("Average Speed of Answer Over Time")
plt.show()
```

# Scatter plot

```python
plt.scatter(df["AvgTalkDuration(sec)"], df["Speed of answer in seconds"])
plt.xlabel("Avg Talk Duration(sec)")
plt.ylabel("Speed of Answer")
plt.title("Speed of Answer vs Avg Talk Duration")
plt.show()
```

# Histogram Using Seaborn

```python
sns.histplot(df["AvgTalkDuration(sec)"], bins=10, kde=True)
plt.xlabel("Avg Talk Duration(sec)")
plt.ylabel("Frequency")
plt.title("Distribution of Avg Talk Duration")
plt.show()
```

# Box plot using seaborn

```python
sns.boxplot(x="Agent", y="Speed of answer in seconds", data=df)
plt.xlabel("Agent")
plt.ylabel("Speed of answer in seconds")
plt.title("Speed of answer in seconds by Agent")
plt.show()
```
