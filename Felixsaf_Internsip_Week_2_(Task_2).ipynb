```python
# import neccessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# dataset file path
path = "/content/drive/MyDrive/dataset/company_sales_data.csv"

# load dataset
df = pd.read_csv(path)
```

```python
# view dataset
df.head()
df.info()
```

# **Exercise 1**

```python
# line plot for total profit by month
df.plot(x="month_number", y="total_profit", kind="line", marker='o')
plt.title('Total Profit Over Months')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

# Exercise 2

```python
# Sublot
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# line plot for bathing soap by month
axs[0].plot(df['month_number'], df['bathingsoap'], marker='o', label='Bathing Soap')
axs[0].set_title('Bathing Soap Sales Over Months')
axs[0].set_xlabel('Month')
axs[0].set_ylabel('Sales')
axs[0].grid(True)
axs[0].legend()

# line plot for facewash by month
axs[1].plot(df['month_number'], df['facewash'], marker='o', label='facewash')
axs[1].set_title('Facewash Sales Over Months')
axs[1].set_xlabel('Month')
axs[1].set_ylabel('Facewash Sales')
axs[1].grid(True)
axs[1].legend()
```
